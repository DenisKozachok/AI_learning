```python
import dlib
import matplotlib.pyplot as plt
import cv2  # OpenCV for image manipulation
```

### Step 1

Load an image (any image that contains faces).

```python
image_path = 'D:\AI_learning\images\zelensky.jpg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
```

### Step 2

Load the dlib face predictor.

```python
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
detected_faces = detector(image_rgb, 1)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
border_color = (255, 255, 0) #    BGR

for i, face_rect in enumerate(detected_faces):
    x1, y1, x2, y2 = face_rect.left(), face_rect.top(), face_rect.right(), face_rect.bottom()
    cv2.rectangle(image_rgb, (x1, y1), (x2, y2), border_color, 2)

plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
